## Transferring Data from Jupyter ro Workbenck

So let's create the query and then store its first part in a variable called insert underscore query. In other words we will assign the string insert into predicted output values to it thus indicating we are getting ready to insert some information.

How can this operation be programmed by writing loops that will iterate over both the rows and columns.Which means that when we are on the first row we'd like to extract the value contained in its first column then each subsequent column until the last the same process must then be repeated for all rows as the NUM PIs shape method tells us the iteration will go over 13 columns and 40 rows then we must find a way to extract a single value from our table.

````python
#it bring the information from 6. column which is the age column
print(df_new_obs[df_new_obs.columns.values[6]].head())
````

````python
#if we want to bring the 6. column first observation we are going to use the following information.
print(df_new_obs[df_new_obs.columns.values[6]][0])
````

````python
#  create the query 
insert_query = 'INSERT INTO predicted_outputs VALUES'
````

````python
insert_query
````

````python
# the loop that will complete the insert query.
#we want our iteration to go through every row
for i in range(df_new_obs.shape[0]):
    #we want to start that code of every record with an open parentheses
    insert_query += '('
    #for inner loop for every column of our data frame 
    for j in range(df_new_obs.shape[1]):
        #we went to add the values from all columns corresponding to the given row separating them with a comma
        insert_query += str(df_new_obs[df_new_obs.columns.values[j]][i]) + ','
    #The insert query syntax requires each row  to end with parentheses not with a comma
    #the insert query by practically the same variable without the last two characters 
    #the comma in the whitespace then it replaces these two characters with a close parenthesis and a comma
    insert_query = insert_query[:-1] + '),'
````

````python
insert_query
````

This is the exact MySQL will insert statement we needed displaying as a string variable in Python with one little flaw though. Compare the last symbol used in the two queries are Python string doesn't end with a proper statement
`Terminator which is semicolon`. That's why we will correct the end of our insert query once more this time we'll do this outside the loop.

````python
insert_query = insert_query[:-1]+';'
````

````python
insert_query
````

````python
# to transfer this table to the workbench we need to  apply the execute method
cursor.execute(insert_query)
````

````python
#When you're executing MySQL statement affecting the database state from Python you must always commit to this action
conn.commit()
````
